```python
rangos = diabetes.agg(['min', 'max']).T.rename(columns={'min':'Min','max':'Max'})
rangos
```

```python
# Promedio, mediana, desviacion estandar, minimo, y maximo.
resumen = pd.DataFrame({
    "count": diabetes.count(),
    "mean": diabetes.mean(numeric_only=True),
    "median": diabetes.median(numeric_only=True),
    "std": diabetes.std(numeric_only=True),
    "min": diabetes.min(numeric_only=True),
    "max": diabetes.max(numeric_only=True)
}).round(2)
resumen
```
