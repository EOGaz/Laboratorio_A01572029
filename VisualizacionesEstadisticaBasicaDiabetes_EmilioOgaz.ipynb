# Visualizacion de Datos

```python
# Crea un subconjunto (comentario del PDF) y cálculo de medias por grupo ya realizado arriba
diagnostico_clase = diagnostico.groupby('Outcome')['Glucose'].mean()
```

```python
plt.hist(diabetes['Age'], bins=4)
plt.title('Histograma de Edad')
plt.show()
```

La distribución está concentrada entre 20 y 35 años, con una caída progresiva a partir de los 40. Se muestran muy pocos casos mayores de 60–80 años. Esto demuestra una mayor cantidad de adultos jóvenes, con presencia baja de personas de edad avanzada.

```python
plt.hist(diabetes['Glucose'], bins=4)
plt.title('Histograma de Glucosa')
plt.show()
```

La mayor densidad aparece alrededor de 100–140 mg/dL, con una barra derecha que alcanza 180–200 mg/dL. Hay valores extremos muy bajos (cercanos a 0).

```python
plt.scatter(diabetes['Age'], diabetes['Glucose'])
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.title('Edad vs Glucosa')
plt.show()
```

El scatter de puntos no muestra una relación tal cual fuerte. Personas de 20–40 años abarcan casi todo el rango de glucosa, y en edades mayores sigue habiendo gran dispersión. Es decir, la glucosa elevada puede aparecer a distintas edades sin un patrón claro. Probable signifique que se registraron mas casos de glucosa alta con adultos jovenes.

# Boxplot

```python
#Boxplot de edad
print("Mediana Edad:", diabetes['Age'].median())
print("Media Edad:", diabetes['Age'].mean())
print(diabetes['Age'].dtype)
diabetes["Age"].isnull().sum()

plt.boxplot(diabetes['Age'])
plt.title('Boxplot de Edad')
plt.show()
```

La mediana se ubica cerca de los 29 años, con un rango aproximado de 23–41. Los bigotes se extienden hacia valores altos y se observan varios por encima de ~65–80, lo que confirma la asimetría y la presencia de unos pocos individuos mayores que la mayoría.

```python
#Usando seaborn
#import seaborn as sns
sns.boxplot(diabetes['Glucose'])
plt.title('Boxplot de Glucosa')
plt.show()
```

Mediana cercana a 120 mg/dL y un rango que ronda 100–140. El diagrama evidencia una variabilidad marcada y cierta asimetría, que se relaciona con el histograma.

```python
#Boxplot de edad de las personas que sobreivivieron y no sobrevivieron
sns.boxplot(diabetes, x="Glucose", y="Age")
sns.stripplot(diabetes, x="Glucose", y="Age", edgecolor="auto" )
plt.title("Edad de acuerdo a Glucosa")
```

Se grafica la distribución de la edad para cada valor específico de glucosa. Como Glucose es continua, aparecen muchas “categorías” estrechas en el eje x y el gráfico se ve muy saturado. Aun así, se ve que la mediana de edad no cambia drásticamente entre niveles de glucosa y la dispersión es amplia en casi todo el rango.

# Matriz de Correlacion

```python
#Seleccionar variables numericas
variables_numericas = diabetes.select_dtypes(include='number')
matriz_correlacion = variables_numericas.corr().round(2)
matriz_correlacion
```

# Mapa de Color

```python
#Mostrar mapa de calor.
sns.heatmap(matriz_correlacion, annot=True)
plt.title('Correlación entre características "')
```

Resalta correlaciones moderadas como Edad a Embarazos (0.54) y Edad a FamilySize (0.56), además de Glucosa a Outcome (0.47). La relación casi perfecta entre Embarazos y FamilySize (0.99) es esperable porque FamilySize se construyó a partir de esas variables. La mayoría de las demás asociaciones son débiles, con algunos valores negativos leves y muchos cercanos a cero.

```python
sns.heatmap(matriz_correlacion,annot=True,cmap='YlGnBu',linecolor="white")
plt.title('Correlación entre características"')
```

Presenta la misma matriz con otro esquema de color que facilita distinguir magnitudes intermedias.
